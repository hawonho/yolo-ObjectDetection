```python
import numpy as np
```

```python
img_rgb_copy = img_rgb.copy()
ground_truth_box = [60, 15, 320, 420]
```

```python
red_rgb = (255, 0, 0)
```

```python
img_rgb_copy = cv2.rectangle(
    img_rgb_copy,
    (ground_truth_box[0], ground_truth_box[1]),
    (ground_truth_box[2], ground_truth_box[3]),
    color=red_rgb, thickness=2
)
```

```python
green_rgb = (0, 255, 0)
```

```python
for rect in cand_rects:
    left = rect[0]
    top = rect[1]
    right = left + rect[2]
    bottom = top + rect[3]
    x1 = np.maximum(left, ground_truth_box[0])
    y1 = np.maximum(left, ground_truth_box[1])
    x2 = np.maximum(left, ground_truth_box[2])
    y2 = np.maximum(left, ground_truth_box[3])
    intersection = np.maximum(x2 - x1, 0) * np.maximum(y2 - y1, 0)
    cand_box_area = (right - left) + (bottom - top)
    gt_box_area = (ground_truth_box[2] - ground_truth_box[0]) * (ground_truth_box[3] - ground_truth_box[1])
    union = cand_box_area + gt_box_area - intersection
    iou = intersection / union
    print("iou = ", iou)
    if iou > 0.5 :
        img_rgb_copy = cv2.rectangle(img_rgb_copy, (left, top), (right, bottom), color=green_rgb, thickness=2)
        cv2.putText(img_rgb_copy, str(iou), (rect[0]+ 100, rect[1]+10), cv2.FONT_HERSHEY_SIMPLEX, 1, color=green_rgb, thickness=2)
```
